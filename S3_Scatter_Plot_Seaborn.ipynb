## **Importing Libraries**

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

##**Loading Datasets**

```python
df = pd.read_csv('/content/drive/MyDrive/Data-Sciecne(Funcdamentals)/S_Seaborn_Data_Science/titanic.csv')
df
```

##**Scatter-Plot**

```python
sns.scatterplot(x='age', y='fare', data=df)
```

```python
sns.scatterplot(x = 'age'  , y = 'fare' , data = df , hue = 'sex')
```

```python
sns.scatterplot(x = 'age'  , y = 'fare' , data = df , hue = 'alive')
```

```python
sns.scatterplot(x = 'age'  , y = 'fare' , data = df , hue = 'alive' , style = 'class' , palette = 'deep')
plt.title('Titanic Data Analysis')
plt.figure(figsize = (12 , 5))

# Seaborn Default Color Palette
# Not only does this function allow you the ability to use Seaborn default colors, but also any of Seaborn's other styling techniques. Seaborn has six variations of its default color palette: deep , muted , pastel , bright , dark , and colorblind .

```

```python
sns.scatterplot(x = 'age'  , y = 'fare' , data = df , hue = 'alive' , style = 'class' , palette = 'gist_rainbow'
, alpha = 0.5)
plt.title('Titanic Data Analysis')
plt.figure(figsize = (12 , 5))
```

```python
sns.scatterplot(x = 'age' , y = 'fare' , data = df , hue = 'alive' , style = 'class')
sns.lineplot(x = 'age' , y = 'fare' , data = df , color = 'green')
plt.figure(figsize = (40 , 39))
```

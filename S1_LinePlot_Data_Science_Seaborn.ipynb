## **Importing Libraris**

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

##**Loading Datasets**

```python
df = pd.read_csv('/content/sample_data/california_housing_test.csv')
df
```

```python
df2 = df[['households' , 'population']]
df2
```

##**Line Plot**
